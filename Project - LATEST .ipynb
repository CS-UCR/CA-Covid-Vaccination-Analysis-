```python
import requests
response = requests.get("https://www.california-demographics.com/counties_by_population")

if response.status_code == 200:
    print('all good')
else:
    print('error')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")
```

```python
headings = soup.find_all(['tr'])
```

```python
rows = []


for name in headings[1:-1]:
    county = name.find('a').text
    population = name.find('td').find_next('td').find_next('td').text
    
    population = population.replace('\n', '')
    population = population.replace(' ', '')
    population = population.replace(',', '')
   
    
    
   
    rows.append({
        "California Counties": county,
        "Population": int(population)
    })
```

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

df_pop = pd.DataFrame(rows)
#df_pop
```

```python
df_pop.sort_values('California Counties', ascending=True, inplace = True)
#df_pop
```
