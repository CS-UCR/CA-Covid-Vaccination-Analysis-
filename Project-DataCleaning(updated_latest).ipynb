```python
import pandas as pd
#pd.to_datetime('2020-12-15 3:45pm')
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv("CA_Vaccine_Data.csv") 
df
```

```python
# DATA CLEANING:-
# RENAME ALL THE COLUMNS TO PROPERLY VISUALIZE!
```

```python
df.rename(columns={'county':'California Counties','administered_date':'Dose Date YYY-DD-MM', 'total_doses':'Total Doses', 'cumulative_total_doses':'Cumulative Total',
                   'pfizer_doses': 'PFIZER', 'cumulative_pfizer_doses':'PFIZER TOTAL','moderna_doses':'MODERNA','cumulative_moderna_doses':'MODERNA TOTAL',
                     'jj_doses':'JANSSEN','cumulative_jj_doses':'JANSSEN TOTAL','partially_vaccinated':'Partially Vaccinated',
                       'total_partially_vaccinated':'TOTAL Partially Vaccinated','fully_vaccinated':'Fully Vaccinated','cumulative_fully_vaccinated':'TOTAL Fully Vaccinated',
                         'at_least_one_dose':'One Dose Taken','cumulative_at_least_one_dose':'TOTAL One Doses Taken'}, inplace = True)
df
```

```python
df.drop(["california_flag"],axis=1, inplace = True)
df.set_index('California Counties', inplace = True)
df.drop(labels = ["Unknown", "All CA and Non-CA Counties", "Outside California"], axis=0, inplace = True)
df.reset_index(inplace = True)
df
```

```python
for col in df:
    print(df[col].unique())
```

```python
import matplotlib
%matplotlib inline
```

```python
df.groupby('Dose Date YYY-DD-MM')[['PFIZER TOTAL', 'MODERNA TOTAL', 'JANSSEN TOTAL']].mean().plot.hist(bins=70)
```

```python
# Checking the data type 
df.dtypes
```

```python
# Setting 'Dose Date' as an index 
df = df.set_index('Dose Date YYY-DD-MM')
df.index
```

```python
# Number of PFIZER, MODERNA, JANSSEN Doses taken on particular dates
cols_plot = ['PFIZER', 'MODERNA', 'JANSSEN']
axes = df[cols_plot].plot(marker='.', alpha=1, linestyle='None', figsize=(11, 9), subplots=True)
for ax in axes:
    ax.set_ylabel('Doses Taken')
```

```python
# Comparing how many people are fully vaccinated in Riverside and Los Angeleous counties. 
df.groupby(["California Counties"])["Fully Vaccinated"].mean().loc[["Riverside", "Los Angeles"]].plot.pie()
```

```python
import random as rand
from pandas.plotting import scatter_matrix
# print (df.shape)
x = df.iloc[1:17] 
a = scatter_matrix(x, alpha=1, figsize=(20,20), diagonal='hist')
plt.show()
```

```python
# Here we are comparing four CA counties to see which vaccine people have taken the most regardless of dose count. 
df.groupby(["California Counties"])[["PFIZER","MODERNA","JANSSEN"]].mean().loc[["Riverside", "Los Angeles", "Orange", "San Diego"]].plot.bar()
```
